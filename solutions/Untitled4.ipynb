```python
from tkinter import *
import threading
import os
import time
demo_mode = True

def thead_ops(file='some_file'):   
    print(f"File exists? {os.path.exists(file)}\n")
    
    time.sleep(1000)


class Application(Frame):
    global thread
    thread = threading.Thread(target=thead_ops)
    thread.start()
    print(f"Thead alive? {thread.is_alive()}\n")
    pass


root = Tk()
app = Application(master=root)

if demo_mode:
    app.mainloop()
print("DONE!")
print(f"Thead alive? {thread.is_alive()}\n")
print(dir(thread._stop))
thread._stop.set()
print(f"Thead alive? {thread.is_alive()}\n")
```
